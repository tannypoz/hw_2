```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

2 пункт - Визуализируйте распределение значений признаков Survived, Pclass, Age, Sex. Parch

```python
data = pd.read_csv('train.csv')
```

```python
# Гистограмма распределения возрастов пассажиров
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'].dropna(), bins=30, kde=True, color='teal')
plt.title('Распределение возрастов пассажиров', fontsize=16)
plt.xlabel('Возраст', fontsize=12)
plt.ylabel('Количество пассажиров', fontsize=12)
plt.show()
```

```python
# Bar Chart распределения выживших и погибших
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=data, hue='Sex', palette='pastel')
plt.title('Выжившие vs Погибшие', fontsize=16)
plt.xlabel('Выжил (1)/Погиб (0)', fontsize=12)
plt.ylabel('Количество пассажиров', fontsize=12)
plt.legend(title='Пол')
plt.show()
```

```python
# Pie Chart распределения классов путешествий
plt.figure(figsize=(8, 6))
data['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('muted'))
plt.title('Процентное соотношение классов пассажиров', fontsize=16)
plt.ylabel('')
plt.show()
```

```python
# График распределения стоимости билетов
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='coolwarm')
plt.title('Стоимость билетов по классам', fontsize=16)
plt.xlabel('Класс', fontsize=12)
plt.ylabel('Цена билета', fontsize=12)
plt.show()
```

```python
# Круговая диаграмма распределения по полу
plt.figure(figsize=(8, 6))
sex_counts = data['Sex'].value_counts()
plt.pie(sex_counts.values, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('muted'))
plt.title('Доля пассажиров по полу', fontsize=16)
plt.axis('equal')  # Чтобы гарантировать круглую форму
plt.show()
```

```python
# Гистограмма распределения признака Parch
plt.figure(figsize=(10, 6))
sns.histplot(data['Parch'], bins=range(max(data['Parch'])+2), color='lightseagreen')
plt.title('Распределение признака Parch', fontsize=16)
plt.xlabel('Количество родителей/детей', fontsize=12)
plt.ylabel('Количество пассажиров', fontsize=12)
plt.show()
```

Медиана равна около 28–30 лет. Это означает, что половина пассажиров была моложе этого возраста, а вторая половина — старше.


Между верхними и нижними границами ящика содержится средняя половина данных (50% пассажиров).
Примерное положение верхних и нижних краев ящика позволяет сказать, что у большинства пассажиров возраст лежит в диапазоне приблизительно от 20 до 38 лет.


Усы (вертикальные линии сверху и снизу ящика):
В примере усы означают, что подавляющее большинство пассажиров имело возраст от ~0 до ~60 лет.


Точки вне усов (потенциальные выбросы):
Эти точки могут указывать на необычные или экстремально высокие/низкие значения возраста.
Имеются незначительное количество маленьких детей и пожилых людей, чьи возраст существенно отличается от основной массы пассажиров

5 пункт - Постройте график типа pie chart для переменных Survived, Pclass, подпишите доли в процентах

```python
# График для переменной Survived
survived_counts = data['Survived'].value_counts()
labels_survived = ['Погибли (0)', 'Выжили (1)']
colors_survived = ['salmon', 'mediumseagreen']

fig, ax = plt.subplots(figsize=(8, 6))
ax.pie(survived_counts, labels=labels_survived, autopct='%1.1f%%',
       shadow=True, startangle=90, colors=colors_survived)
ax.set_title('Выжили или погибли')
plt.show()
```

```python
# График для переменной Pclass
pclass_counts = data['Pclass'].value_counts(sort=False)
labels_pclass = ['1-й класс', '2-й класс', '3-й класс']
colors_pclass = ['gold', 'lightskyblue', 'coral']

fig, ax = plt.subplots(figsize=(8, 6))
ax.pie(pclass_counts, labels=labels_pclass, autopct='%1.1f%%',
       shadow=True, startangle=90, colors=colors_pclass)
ax.set_title('Расклад по классам путешественников')
plt.show()
```

7 пункт - Постройте интерактивный sunburst plot (визуализация иерархических данных) с помощью plotly. На первом уровне иерархии - количество пассажиров в каждом из классов, а на втором количество женщин/мужчин в этом классе

```python
import plotly.express as px
```

```python
# Подготовим данные для диаграммы
sunburst_data = data.groupby(['Pclass', 'Sex']).size().reset_index(name='Counts')
```

```python
# Переименовываем столбцы для удобства
sunburst_data.columns = ['Class', 'Gender', 'NumberOfPeople']
```

```python
# Создаём диаграмму
fig = px.sunburst(
    sunburst_data,
    path=['Class', 'Gender'],
    values='NumberOfPeople',
    color='NumberOfPeople',
    hover_data={'NumberOfPeople': ':,'},  # формат отображения чисел с запятыми-разделителями тысяч
    color_continuous_scale='Blues',
    title='Иерархия пассажиров по классу и полу',
)
```

```python
# Настройка отображения
fig.update_traces(textinfo='label+percent parent')
fig.update_layout(width=800, height=600)
```
